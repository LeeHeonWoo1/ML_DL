### LSTM
입력 데이터 X는 3차원 형태로 만드는데, LSTM 모델에서 3차원 데이터를 입력데이터로 사용하는 이유는 1차원부터 3차원까지의 역할이 각각 샘플 수, 스텝 수, 기능 수에 해당합니다. LSTM 모델이 각 샘플의 시간 경과에 따른 일련의 입력 값으로 구성되는 시퀀스 데이터와 함게 작동되도록 설계되었기 때문이다.

입력, 결과 데이터를 np.array로 바꾸는 이유는 python의 기본 list보다 Numpy의 배열이 LSTM같은 딥러닝 모델에서 대규모 데이터셋을 처리할 때 더 효율적이기 때문이다.

```python
import numpy as np
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import LSTM, Dense

# 학습 데이터 생성(입력 데이터 X, 결과 데이터 Y)
X = []
Y = []
for i in range(5, 20):
    X.append([[j] for j in range(i-5, i)])
    Y.append([i])

# LSTM 모델 구성
# LSTM 레이어 50개층, Dense 레이어 1개층으로 구성된 모델
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape = (5, 1)))
model.add(Dense(1, activation='linear'))
model.compile(loss = "mse", optimizer="adam")

# 모델 학습, epochs = 100은 100회 학습을 의미,
model.fit(X, Y, epochs=100, batch_size=1)

# 다음 값을 예측한다.
next_value = np.array([[[1], [2], [3], [4], [5]]])
predicted_value = model.predict(next_value)
print(f"predict value : {predicted_value[0][0]}")
```
