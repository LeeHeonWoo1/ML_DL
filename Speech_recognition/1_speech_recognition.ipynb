해당 인스턴스에서는 아래 7가지의 API들로 오디오 소스의 음성을 인식한다. 각 기능에 대한 설명은 링크를 참조하자.

|종류|설명|링크|
|---|---|---|
|recogize_sphinx|CMU Sphinx <br> (PocketSpnix 설치 필요)|https://cmusphinx.github.io/|
|recognize_bing()|MS의 Bing speech|https://azure.microsoft.com/en-us/services/cognitive-services/speech-services/|
|recognize_google()|google의 web speech API|https://wicg.github.io/speech-api/|
|recognize_google_cloud()|google의 cloud speech <br> (google-cloud-speech 설치 필요)|https://cloud.google.com/speech-to-text|
|recognize_houndify()|SoundHound의 API|https://www.houndify.com/|
|recognize_ibm()|IBM의 Speech To Text|https://www.ibm.com/cloud/watson-speech-to-text|
|recognize_wit()|Wit|https://wit.ai/|

```python
import speech_recognition as sr

Recognizer = sr.Recognizer() #인스턴스 생성
mic = sr.Microphone(device_index=1)
fail_cnt = 0

while True:
    with mic as source: # 마이크에 입력하는 내용대로
        print("음성 입력을 시작합니다")
        audio = Recognizer.listen(source, timeout=5)
    try:
        data = Recognizer.recognize_google(audio, language="ko-KR")
        print(f"입력된 음성 : {data}") # 해당 내용을 출력
    except KeyboardInterrupt:
        print("프로그램 종료")
    except:
        print("이해하지 못했음")
        fail_cnt += 1

    if fail_cnt >= 10:
        print("10회 인식 오류로 인해 녹음을 종료합니다.")
        break
        
```

```python
import requests
import speech_recognition as sr

def sendCmd(cmd: str): # 함수는 string형태의 인자값을 받는다.
    esp_ip = "http://192.168.4.1" # ip address
    action_uri = '{}:80/action'.format(esp_ip) # port지정. 80은 http port를 의미 (제어신호 송신용)   
    
    data = {'go': cmd} # 인자값으로 받은 string을 dict형태로 구성하여
    requests.get(action_uri, params=data) # request를 통해 전달한다.

try:
    while True:
        r = sr.Recognizer()
        with sr.Microphone(device_index=1) as source:
            print("음성 입력 중입니다")
            audio = r.listen(source)
            try:
                data = r.recognize_sphinx(audio, language='en-US')
                print(data)
                if "r" in data:
                    print('전진하겠습니다.')
                    sendCmd("forward")

                elif "'" in data:
                    print('후진하겠습니다.')
                    sendCmd("backward")

                elif "f" in data:
                    print('좌회전 하겠습니다.')
                    sendCmd("left")

                elif "what" in data:
                    print('우회전 하겠습니다.')
                    sendCmd("right")  
                    
                else:
                    print("멈추겠습니다.")
                    sendCmd("stop")
            except sr.UnknownValueError:
                print("이해하지 못했습니다.")
            except sr.RequestError:
                print("연결에 문제가 발생했습니다.")
except KeyboardInterrupt:
    pass
```
