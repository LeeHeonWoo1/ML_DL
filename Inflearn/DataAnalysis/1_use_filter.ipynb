## Q1.

```python
Q1 = answer['Q1'].value_counts().sort_index()
sns.countplot(data = answer.sort_values('Q1'), x="Q1", palette="Blues_r").set_title(question['Q1'])
```

```python
def show_countplot_by_Qnum(q_num, figsize = (10, 6), order=None):
    if not order:
        order = answer[q_num].value_counts().index
        
    plt.figure(figsize=figsize)
    sns.countplot(data = answer, y = q_num, palette="Blues_r", order=order).set_title(question[q_num])

show_countplot_by_Qnum("Q6")
```
